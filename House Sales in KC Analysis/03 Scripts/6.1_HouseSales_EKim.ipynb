```python
#to remove only 'T000000' in column name 'date' (e.g. from '20141013T000000' to '20141013')
df_haus['date'] = df_haus['date'].str.replace('T000000', '')
```

```python
#dropping unnecessary columns for this analysis
df_haus = df_haus.drop(columns = ['sqft_above'])
df_haus = df_haus.drop(columns = ['sqft_basement'])
df_haus = df_haus.drop(columns = ['lat'])
df_haus = df_haus.drop(columns = ['long'])
df_haus = df_haus.drop(columns = ['sqft_living15'])
df_haus = df_haus.drop(columns = ['sqft_lot15'])
```
